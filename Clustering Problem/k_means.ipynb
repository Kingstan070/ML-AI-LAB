## K-Means Clustering

```python
# Reading CSV file 
# --- OR -- in this case
# Loading pre-exsisting dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# Printing information of data
print(iris.DESCR)
```

```python
# Getting X and y Values
X = iris.data
y = iris.target
```

```python
# Creating Model
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state= 1)

# Fitting/training the dataset
model.fit(X)
```

```python
# Predicting the clusters using data
pred = model.predict(X)
```

```python
## Ploting the graph
import matplotlib.pyplot as plt
import matplotlib.cm as cm

plt.figure(figsize=(18,5)) # Size of the figure

plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c = pred, cmap=cm.Accent) ## Scattering features at column 0 and 1
for center in model.cluster_centers_:
    ## Scattering center of column 0 and 2
    center = center[0:2]
    plt.scatter(center[0], center[1], marker='^', c='red')
plt.grid(True)
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")

plt.subplot(1,2,2)
plt.scatter(X[:,2],X[:,3],c = pred, cmap=cm.Accent) ## Scattering features at column 2 and 3
for center in model.cluster_centers_:
    ## Scattering center of column 2 and 3
    center = center[2:4]
    plt.scatter(center[0], center[1], marker='^', c='red')
plt.grid(True)
plt.xlabel("Sepal length (cm)")
plt.ylabel("Sepal width (cm)")

plt.show()
```

```python
## Evaluating the Model
from sklearn.metrics import accuracy_score, precision_score, recall_score
ac = accuracy_score(y, pred)
pr = precision_score(y, pred, average='weighted')
re = recall_score(y, pred, average='weighted')
print('Accuracy Score is:', ac)
print('Precision Score is:', pr)
print('Recall Score is :', re)
```

```python
# Evaluating the Model using Confusion Matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y, pred)
ConfusionMatrixDisplay(confusion_matrix=cm).plot()
```
