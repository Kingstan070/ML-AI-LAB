```python
from sklearn.datasets import load_iris
dataset = load_iris()
dir(dataset)
```

```python
X = dataset.data
y = dataset.target
X, y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train, X_test, y_train, y_test
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
a_score = accuracy_score(y_test, y_pred)
p_score = precision_score(y_test, y_pred, average='weighted')
r_score = recall_score(y_test, y_pred, average='weighted')
cm      = confusion_matrix(y_test, y_pred)
# a_score, p_score, r_score, cm
```

```python
import matplotlib.pyplot as plt
print(f"Accurary Score: {a_score*100:.2f}%")
print(f"Precision Score: {p_score*100:.2f}%")
print(f"Recall Score: {r_score*100:.2f}%")
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm)
cm_display.plot()
plt.show()
```
